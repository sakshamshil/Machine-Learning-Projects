**WORK FLOW**


Heart Data -> Data pre processing -> train test split -> logistic regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
heart_data = pd.read_csv('/content/heart_disease_data.csv')
print(heart_data.head())
```

```python
heart_data.shape
```

```python
heart_data.info()
```

```python
heart_data.isnull().sum()
```

```python
heart_data.describe()
```

```python
heart_data['target'].value_counts()
```

0 -> Defective Heart

1 -> Non Defective Heart

```python
#splitting features and target

X = heart_data.drop(columns = 'target', axis = 1)
Y = heart_data['target']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, stratify = Y, random_state = 2)
```

```python
print(X_train.shape, X_test.shape)
```

Model Training


```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Evaluation of training dataset

```python
train_prediction = model.predict(X_train)
train_acc = accuracy_score(train_prediction, Y_train)
print(train_acc)
```

Evaluation of testing data

```python
test_prediction = model.predict(X_test)
test_acc = accuracy_score(test_prediction, Y_test)
print(test_acc)
```

Predictive System

```python
input_data = [41,0,1,130,204,0,0,172,0,1.4,2,0,2]
input_data_np = np.asarray(input_data).reshape(1,-1)

prediction = model.predict(input_data_np)
print(prediction)

print('Heart Defective' if prediction[0] == 0 else 'Heart Healthy'
)
```
