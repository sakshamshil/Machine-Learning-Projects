**To create a model that predicts if the object is a mine or a rock when Sonar sends sound waves reflected back**

**WORK FLOW**

Sonar Data -> Pre-Processing -> Train Test (Split the data into training cases and Testing cases)
We'll use a Logistic Regression Model because it is really good for binary classification

```python
#dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Processing

```python
#Loading the dataset to a pandas Dataframe
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)  #we put header = None as we don't have any column names available
sonar_data.head()
```

```python
sonar_data.shape    #rows and columns
```

```python
sonar_data.describe()   #shows all the stats for all columns
```

```python
sonar_data[60].value_counts()       #Shows how many rocks and how many miles, basically shows total count of distinct values in 60th column
                                    #Return a Series containing counts of unique values.
```

M --> Mine
R --> Rock

```python
sonar_data.groupby(60).mean()        #This finds the mean value for all the rows for M and R
#A groupby operation involves some combination of splitting the object, applying a function,
#and combining the results. This can be used to group large amounts of data and compute operations on these groups.
```

```python
#Now we have to seperate the data and the Labels = M,R
X = sonar_data.drop(columns=60, axis = 1)        #this drops the last column from the dataframe, axis=1 specifies that we are dropping a colunmn
Y = sonar_data[60]                               #we are storing the labels in Y
```

```python
print(X.head())
print(Y.head())
```

Splitting training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.1, stratify = Y, random_state = 1)
#this order is necessary and should be maintained
```

train_test_split : Split arrays or matrices into random train and test subsets.

test_size = 0.1 : represents the proportion of which should be test data, 10% in this case (should be 0.1 or 0.2)

stratify = Y : we need to have data to be split so that the no of rocks dataset and no of mines data set are almost equal in the training/test dataset. (what does it do exactly tho?)

random_state = 1 : splits the data in a particular order, so if 2 people put random_state = 1, there data will be split similarly.

```python
print (X.shape, X_train.shape, X_test.shape)
```

X of 208 records split into training = 187 and test = 21

**Now we will train our Logistic Regression MODEL**

```python
#creating the model
model = LogisticRegression()
```

```python
#Training the model
model.fit(X_train, Y_train)
```

**MODEL EVALUATION**

Now we will evaluate the accuracy of our model. How?

We will run the model on the data and match the results. Any accuracy >70% is good

```python
#accuracy for training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy*100
```

```python
#accuracy for test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
test_data_accuracy*100     #76%
```
