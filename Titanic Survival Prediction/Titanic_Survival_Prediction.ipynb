**Work Flow**

Data -> Data pre processing -> Data Analysis -> Train Test Split -> Logistic Regression Model -> Evaluation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_data = pd.read_csv('/content/train.csv')
```

```python
titanic_data.head()
```

0 -> Didn't Survive

1 -> Survived

###Handling Missing Values

```python
titanic_data = titanic_data.drop(columns = "Cabin", axis = 1)
#dropping Cabin columns since most of the values are missing
```

```python
#Replacing all the missing values of Age column and replace it with mean of Age
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace = True)
```

```python
titanic_data.head()
```

```python
#Replacing embarked values with mode of the embarked
print(titanic_data['Embarked'].mode())
print(titanic_data['Embarked'].mode()[0])
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace = True)
```

```python
titanic_data.isnull().sum()
```

###Encoding the Categorical Values

```python
titanic_data.replace({"Sex" : {'male' : 0, 'female': 1}, "Embarked" : {'S' : 0, 'C' : 1, 'Q' : 2}}, inplace = True)
```

```python
titanic_data.head()
```

```python
#since name, PassengerID, Ticket is irrelevant

#Seperating features and labels

X = titanic_data.drop(columns = ['Name', 'Ticket' , 'PassengerId',  'Survived'], axis = 1)
Y = titanic_data['Survived']
```

```python
print(X)
```

Splitting into Training and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

##Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

##Evaluation

```python
training_prediction = model.predict(X_train)
training_accuracy = accuracy_score(training_prediction, Y_train)
print(training_accuracy)
```

```python
testing_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(testing_prediction, Y_test)
print(testing_accuracy)
```

###Building a Predictive Model

```python
input_data = [3,1,4,1,1,16.7,0]
input_np = np.asarray(input_data).reshape(1,-1)

prediction = model.predict(input_np)

print(prediction)

print("Patient Survived" if prediction[0] == 1 else "Person didn't survive")
```
